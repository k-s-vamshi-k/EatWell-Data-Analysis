```python
import folium
from geopy.geocoders import Nominatim
from folium.plugins import MarkerCluster

top_locations = df['location'].value_counts().head(10).index

df_top_locations = df[df['location'].isin(top_locations)]
top_restaurants_in_top_locations = df_top_locations['name'].value_counts().head(10)

geolocator = Nominatim(user_agent="my_geocoder")

geocoded_locations = {}
for location in top_locations:
    location_data = geolocator.geocode(location)
    if location_data:
        geocoded_locations[location] = (location_data.latitude, location_data.longitude)


df_top_locations['latitude'] = df_top_locations['location'].map(lambda x: geocoded_locations[x][0] if x in geocoded_locations else None)
df_top_locations['longitude'] = df_top_locations['location'].map(lambda x: geocoded_locations[x][1] if x in geocoded_locations else None)

map_restaurants = folium.Map(location=[df_top_locations['latitude'].mean(), df_top_locations['longitude'].mean()], zoom_start=12)

marker_cluster = MarkerCluster().add_to(map_restaurants)

for index, row in df_top_locations.iterrows():
    if row['name'] in top_restaurants_in_top_locations.index:
        folium.Marker([row['latitude'], row['longitude']], popup=row['name']).add_to(marker_cluster)

for location in top_locations:
    folium.Marker([geocoded_locations[location][0], geocoded_locations[location][1]], popup=location).add_to(map_restaurants)

map_restaurants
```

* The map gives the location of top Restaurants in Banglore.

## Top Rated Restaurants

```python
import re
def clean_name(name):
    cleaned_name = re.sub(r'[^a-zA-Z0-9\s]', '', name)
    return cleaned_name.strip()
df['Name of the Restaurant'] = df['name'].apply(clean_name)
avg_ratings = df.groupby('Name of the Restaurant')['rate'].mean().sort_values(ascending=False).round(2)
top_rated_restaurants = avg_ratings.head(10)
top_rated_restaurants = pd.DataFrame(top_rated_restaurants)
top_rated_restaurants = top_rated_restaurants.rename(columns = {'rate':'Average Ratings'})
top_rated_restaurants
```

## Average cost for two in Top 10 Locations
